# Bengaluru Fitness Centre Market Analysis


## Objective

This project analyses fitness centres across Bengaluru to identify high-opportunity locations for launching a fitness-related business targeting students and working professionals aged 20–40.


## Dataset Summary

- Total Records: 500
- City: Bengaluru
- Categories Covered:
  - Gym
  - Yoga
  - Zumba
  - Pilates
  - CrossFit
  - Kickboxing
  - Karate
  - Taekwondo

## Tools Used

- Python
- Pandas
- Matplotlib
- VS Code

## Data Cleaning Steps


- Removed duplicates
- Standardized categories
- Handled missing values
- Exported cleaned CSV and Excel files

# Analysis & Insights

## 1. Rating Distribution
Most fitness centres have ratings between 3.5 and 4.5, indicating generally positive customer satisfaction.

## 2. Category Analysis
Gym and Fitness Studios are the most common fitness category in Bengaluru.

## 3. Top Localities
Areas like Indiranagar, Koramangala, and Whitefield show high concentrations of fitness centres.

## 4. Reviews vs Ratings
Fitness centres with higher reviews generally maintain better ratings, showing trust and popularity.

# Business Recommendation

## Best Area to Launch
Indiranagar and Koramangala appear to be strong business opportunities due to high youth population and fitness awareness.

## Underserved Opportunity
Outer Whitefield and Electronic City show moderate competition and good expansion potential.

## Best Category Opportunity
Pilates and CrossFit studios appear less saturated compared to traditional gyms.

## Target Audience
Young professionals and students aged 20–40 are highly concentrated near metro-connected areas and IT hubs.
